If you have 8 choices of meat and 9 choices of sides, how many possible arrangement of meats and sides are there if you get to pick two of each? Repetition is allowed and order DOESN'T matter.

Repetition, order doesnt matter: $\frac{(N - 1 + n)!}{n!(N-1)!}$



The answer is $36 * 45 = 1620$ as shown below.

```python
import itertools
```

```python
meats = ['Georgia Chopped Pork','Country-Roasted Chicken','Barbeque Chicken',
         'Texas Beef Brisket','Barbeque Pulled Chicken','Southside Rip Tips',
         'Traditional Bar-B-Que Wings','Hot Link Sausage']
```

```python
sides = ['Creamy Coleslaw','Wilbur Beans','Potato Salad','Daves Cheesy Mac & Cheese',
         'Carlic Red-Skin Mashed Potatoes','Sweet Corn','Famous Apples','Collard Greens',
         'Corn Bread Muffin']
```

```python
#All combinations of two meats and two sides (repetitions, order DOESNT matter)
all_combinations_meat = list(itertools.combinations_with_replacement(meats,2))
all_combinations_sides = list(itertools.combinations_with_replacement(sides,2))
all_comb_meats_sides = list(itertools.product(all_combinations_meat,all_combinations_sides))
```
